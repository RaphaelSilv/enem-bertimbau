# Helper functions

```python
def get_tema(row):
  tema = ""

  try:
    # print(f"{row} - {prompts[prompts['id'] == row].category.item()}")
    tema = prompts[prompts['id'] == row].title.item()
  except:
    # print(f"{row} - {None}" )
    tema = None
  return tema

def get_texto_motivador(row):
  texto_motivador = ""

  try:
    # print(f"{row} - {prompts[prompts['id'] == row].category.item()}")
    texto_motivador = prompts[prompts['id'] == row].description.item()
  except:
    # print(f"{row} - {None}" )
    texto_motivador = None
  return texto_motivador


def get_categoria(row):
  categoria = ""

  try:
    # print(f"{row} - {prompts[prompts['id'] == row].category.item()}")
    categoria = prompts[prompts['id'] == row].category.item()
  except:
    # print(f"{row} - {None}" )
    categoria = None
  return categoria
```

```python
import re

def limpa_doc(document, encoding=None):

  CARACTERES_INDESEJADOS="“”[\\]^<>`´{|}'~—\""

  try:

    if isinstance(document, list):
      document = "".join(document)

    table = document.maketrans('', '', CARACTERES_INDESEJADOS)
    document = document.translate(table).strip()

  except:
    print('exception')
    document = None

  return document


def limpa_texto_motivador(texto, encoding=None):
    texto_limpo = ""

    sents_to_remove =    [
        "Texto I:",
        "Texto II:",
        "Texto III:",
        "Texto i:",
         "Texto ii:",
        "Texto iii:",
        "Texto 1:",
        "Texto 2:",
        "Texto 3:",
        "Texto 4:",
        "(leia sobre a importância do caráter ‘único’ do SUS no enfrentamento da epidemia na próxima reportagem desta série)",
        "(traduzido e adaptado)",
        "(editado e adaptado)",
        "G1 - Economia (Editado e adaptado)",
        "[Leia na íntegra - Portal UOL]",
        "[Portal ESPN - leia na íntegra]",
        "[Revista Superinteressante - leia na íntegra]",
        "[Blog do Planalto - leia na íntegra]",
        "[Portal JusBrasil - leia na íntegra]",
        "[ Le Monde Diplomatique Brasil - Leia na íntegra]",
        "[ Leia na íntegra - Portal EBC]",
        "UOL Educação",
        "BBC Brasil, editado",
        "UOL Notícias (Editado)",
        "UOL Leia o texto na íntegra",
        "Wikipédia Leia o texto na íntegra",
        "Congresso em Foco Leia o texto na íntegra",
        "Fiocruz Leia o texto na íntegra",
        "G1 Leia o texto na íntegra",
        "Veja Leia o texto na íntegra",
        "Brasil Escola Leia o texto na íntegra",
        "MEC Leia o texto na íntegra",
        "Farming Brasil Leia o texto na íntegra",
        "Correio Braziliense Leia o texto na íntegra",
        "Outras Palavras Leia o texto na íntegra",
        "Portal ESPN - leia na íntegra",
        "Revista Superinteressante - leia na íntegra",
        "Folha de São Paulo (editado)",
        "Folha de S. Paulo (editado)"
        "(site da ISTOÉ, editado)",
        "Leia o texto na íntegra"
    ]

    patterns = [re.compile(re.escape(pattern)) for pattern in sents_to_remove]

    if isinstance(texto, list):
        texto_limpo = limpa_doc("".join(texto))
    else:
        texto_limpo = limpa_doc(texto)

    for pattern in patterns:
      # print(f"p: {pattern}")
      if re.search(pattern, texto_limpo):
        # print(f"pattern: {pattern}")
        texto_limpo = re.sub(pattern, "", texto_limpo)
        # print(f"texto: {texto_limpo}")
    return texto_limpo
```

```python
from transformers import AutoTokenizer  # Or BertTokenizer
from transformers import AutoModelForPreTraining  # Or BertForPreTraining for loading pretraining heads
from transformers import AutoModel  # or BertModel, for BERT without pretraining heads

model = AutoModelForPreTraining.from_pretrained('neuralmind/bert-base-portuguese-cased')
tokenizer = AutoTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased', do_lower_case=True)
```

```python
import torch

def get_embeddings_from_text(text):
  encoded_input = tokenizer(text, return_tensors='pt', truncation=False)

  with torch.no_grad():
    outs = model(**encoded_input)
    encoded = outs[0][0, 1:-1] # Ignore [CLS] and [SEP] special tokens

  return encoded
```

```python
import spacy

nlp = spacy.load('pt_core_news_lg')


def split_text_into_sentences(texto):
    doc = nlp(texto)
    try:
      return [sent.text for sent in doc.sents]
    except:
      return None

def get_BERT_tokens_from_text(text):
    return tokenizer.convert_ids_to_tokens(tokenizer.encode(text))
```

# Gera coluna de sentenças utilizando Spacy





```python
df['sentencas_redacao']=df.progress_apply(lambda row: split_text_into_sentences(row['redacao']), axis=1)
df['sentencas_texto_motivador']=df.progress_apply(lambda row: split_text_into_sentences(row['texto_motivador']), axis=1)
df['qtd_sentencas_redacao']=df.progress_apply(lambda row: len(row['sentencas_redacao']), axis=1)
df['qtd_sentencas_texto_motivador']=df.progress_apply(lambda row: len(row['sentencas_texto_motivador']), axis=1)
```

# Gera colunas de BERT tokens por sentenças e por texto inteiro

```python
df['redacao_bert_tokens']=df.progress_apply(lambda row: get_BERT_tokens_from_text(row['redacao']), axis=1)
df['texto_motivador_bert_tokens']=df.progress_apply(lambda row: get_BERT_tokens_from_text(row['texto_motivador']), axis=1)
```

```python
df['len_redacao']=df.progress_apply(lambda row: len(row['redacao']), axis=1)
df['len_texto_motivador']=df.progress_apply(lambda row: len(row['texto_motivador']), axis=1)
```

```python
df['qtd_bert_tokens_redacao']=df.progress_apply(lambda row: len(row['redacao_bert_tokens']), axis=1)
df['qtd_bert_texto_motivador']=df.progress_apply(lambda row: len(row['texto_motivador_bert_tokens']), axis=1)
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df['tokens_sentencas_redacao']=df.progress_apply(lambda row: [get_BERT_tokens_from_text(sent) for sent in row['sentencas_redacao']], axis=1)
df['tokens_sentencas_texto_motivador']=df.progress_apply(lambda row: [get_BERT_tokens_from_text(sent) for sent in row['sentencas_texto_motivador']], axis=1)
df['tokens_sentencas_tema']=df.progress_apply(lambda row: get_BERT_tokens_from_text(row['tema']), axis=1)
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df = df[df['qtd_sentencas_redacao'] > 0]
```

# Gera média dos tamanhos de sentenças das redações e textos motivadores

```python
df['avg_len_sentencas_redacao'] = df.progress_apply(lambda x: x['len_redacao'] / x['qtd_sentencas_redacao'], axis=1)
df['avg_len_sentencas_texto_motivador'] = df.progress_apply(lambda x: x['len_texto_motivador'] / x['qtd_sentencas_texto_motivador'], axis=1)
```

```python
df['avg_bert_tokens_sentencas_texto_motivador'] = df.progress_apply(lambda x: sum(len(s) for s in x['tokens_sentencas_texto_motivador']) / len(x['tokens_sentencas_texto_motivador']), axis=1)
df['avg_bert_tokens_sentencas_redacao'] = df.progress_apply(lambda x: sum(len(s) for s in x['tokens_sentencas_redacao']) / len(x['tokens_sentencas_redacao']), axis=1)
```
